### Intro to DataFrame

```python
import pandas as pd
```

<b> <li>DataFrame or df is an advace series with more than one columns and rows as well (but that is same as series).<br>
<li>Used for data cleaning and analysis.<br>
<li>Or we can say dataframe is a collection of multiple series object.<br>
<li>We use variable.dtype() to view the dtype of a dataframe or series.<br>
<li>We use variable.astype() to change the dtype of a dataframe or series.

```python
# we can pass a list of dictonaries to form a dataframe
r1 = pd.Series(['Alice', 'Physics', 85], index = ['Name', 'Class', 'Score'])
r2 = pd.Series(['Jack', 'Chemistry', 78], index = ['Name', 'Class', 'Score'])
r3 = pd.Series(['Sam', 'English', 92], index = ['Name', 'Class', 'Score'])
```

```python
# With Index
df_record_1 = pd.DataFrame([r1,r2,r3], index = ['School_1','School_2','School_3'])
df_record_1.head()
```

```python
# Without Index
df_record_2 = pd.DataFrame([r1,r2,r3])
df_record_2.head()
```

```python
# Or we could pass a list of dictonaries in our dataframe object
df_exp = pd.DataFrame([{'Name' : 'Alice', 'Subject' : 'Physics','Scores' : 85},
                      {'Name' : 'Jack', 'Subject' : 'Chemistry','Scores' : 78},
                      {'Name' : 'Sam', 'Subject' : 'English','Scores' : 92}])
df_exp.head()
```

```python
# We use loc and iloc the same way use in series but this time it will return a series as dataframe is a collection of series itself
print(df_record_1.iloc[0])  # By index number
print()
print((df_record_1.loc['School_1']))  # By index name
print()
print(type(df_record_1.iloc[0]))  # It is a pandas series datatype
```

```python
# For transpose
df_exp.T   # We can use chaning and use loc and iloc by using df.T.loc[] as it return the transposed Series and DataFrame
```

```python
# Chaning exp in a loc function
print(df_record_1.loc['School_1']['Name'])
print(df_record_1.loc['School_1']['Score'])
```

```python
# If we got confused use the type function
print(type(df_record_1.loc['School_1']))  # Its a series
print(type(df_record_1.loc['School_1']['Name'])) # Its str,int,obj,float,etc
```

<b> <li>Note: Chaning return a copy of the df at the place of a view and the original has no change in it.Whic may slow the computing at first but does not effect the overall performance of the df functions.
<li>For this we can use slicing where we use ( : ) for selecting whole data and list of paticular rows or cols we want in our df's copy.

```python
# Deleting data from a df using drop
df_record_1.drop("School_1")
# Note: It does not delete the data from the original dataframe it just returns a copy as the previous one.
# For permanent deletion from original we use inplace = True which is an optional parameter
# Drop has one more parameter called axis which is by default 0 (rows) but we can make it cols (by putting axis = 1) and this will delete the columns
```

```python
# Exp of axis (Dropping a column)
copy_df = df_record_1.copy() # Make a copy
copy_df.drop("Name", inplace = True, axis = 1)
copy_df # Deleted the Name column permanently
```

```python
# Adding new col just as series
copy_df["Name"] = [1,2,3]
copy_df["Loc"] = None
copy_df
```

#### Manipultion of Data in a DataFrame

```python
df_man = pd.read_csv("presidents.csv")
df_man.head()
```

```python
df_man["First"] =  df_man["President"] # Create a new column named "First" with data of "President" column
# Using replace function we can only assing the first name of the president to first using regex concept
df_man["First"] = df_man["First"].replace("[ ].*","", regex = True)
df_man.head()
```

```python
# Or by using general method of for and split
del (df_man["First"])
def splitname (row):
    row["First"] = row["President"].split(" ")[0]
    row["Last"] = row["President"].split(" ")[-1]
    return row
df_man = df_man.apply(splitname, axis = 1)
df_man.head()
```

```python
# we can also use regex to extract particular data from a row or column using dot str dot extract
df_man["Regex"] = df_man["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df_man["Regex"].head()# extract born date using regex in a requied pattern as we need
```

```python
pd.to_datetime(df_man["Regex"]).head()   # convert argument to datetime  YYYY-MM-DD
```
